## UCI Mechine learning reposiotry : Wisconsin breast cancer data

   1. Sample code number (index) : id number

   2. Clump Thickness : 1 - 10

   3. Uniformity of Cell Size : 1 - 10

   4. Uniformity of Cell Shape : 1 - 10

   5. Marginal Adhesion  : 1 - 10

   6. Single Epithelial Cell Size : 1 - 10

   7. Bare Nuclei : 1 - 10

   8. Bland Chromatin : 1 - 10

   9. Normal Nucleoli : 1 - 10

   10. Mitoses : 1 - 10

   11. Class (label) : 2 is for benign /  4 is for malignant 


reference : https://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+%28original%29                
                   

```python
#UCI mechine learing breast cancer data
data_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data'
df = pd.read_csv(data_path, index_col = 0, parse_dates = True)
df.columns = [ 'clump', 'cell_size', 'cell_shape', 'adhension','epithlial','bare_nuclei','chromatin', 'normal_nucleoli','mitoses','class']

X_data = df.iloc[:, :-1]
y_data = df.iloc[:, -1]

#preprocessing
X_data[X_data['bare_nuclei']=='?']=0 # '?' 값을 0 값으로 바꾸기
X_data['bare_nuclei']=pd.to_numeric(X_data['bare_nuclei']) # X_data['bare_nuclei']의 object 데이터 형식을 int로 바꾸기
```

```python
X_data.info()
```

```python
y_data
```

```python
# tranin dataset,  test dataset 분리
num = int(len(df)*0.8)
train_X, train_y  = X_data.iloc[:num, :], y_data.iloc[:num]
test_X, test_y  = X_data.iloc[num:, :], y_data.iloc[num:]
```

### Decision Tree 학습 및 분류

-default : Decision Tree  (max_depth=10, min_samples_split=2,  min_gain=1e-4)

```python
import sklearn
from sklearn import metrics as metrics
```

```python

print("==============decision tree==============")
# decision tree  학습
tic = time.process_time()
tree = DecisionTree(15,  2, 1e-4)
tree.fit(train_X, train_y)
toc = time.process_time()
print('decision tree 학습에 걸린 시간= {:.4f}초 '.format(toc-tic))

# decision tree  분류
tic = time.process_time()
train_prediction = tree.classify(train_X)
test_prediction = tree.classify(test_X)
toc = time.process_time()
print('decision tree 분류에 걸린 시간= {:.4f}초 '.format(toc-tic))


#train, test accuracy
print('train data accuracy : {:.4f}'.format(metrics.accuracy_score(train_y, train_prediction)) )
print('test data accuracy  : {:.4f}'.format(metrics.accuracy_score(test_y, test_prediction)) )
```


### Random Forest 학습 및 분류

-default : RandomForest (numberOfTrees=20, max_depth=10, min_samples_split=2,  min_gain=1e-4, random_state=100)

```python

print("==============Random Forest==============")
# Random Forest  학습
tic = time.process_time()
rf = RandomForest(20, 15,  2, 1e-4)
rf.fit(train_X, train_y)
toc = time.process_time()
print('Random Forest 학습에 걸린 시간= {:.4f}초 '.format(toc-tic))

# decision tree  분류
tic = time.process_time()
train_prediction = rf.classify(train_X)
test_prediction = rf.classify(test_X)
toc = time.process_time()
print('Random Forest 분류에 걸린 시간= {:.4f}초 '.format(toc-tic))


#train 데이터 정확도
print('train data accuracy : {:.4f}'.format(metrics.accuracy_score(train_y, train_prediction)) )

#test 데이터 정확도 
print('test data accuracy : {:.4f}'.format(metrics.accuracy_score(test_y, test_prediction)) )
```
